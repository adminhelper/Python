# 파이썬의 컬렉션 타입
## 리스트
   [], 순서가 있고 중복가능
## 튜플
   (), 읽기전용 리스트
## 셋
   {}, 집합, 중복불가능
## 딕셔너리
   {key:value}, 맵, 해시

```python
names = ['홍길동', '임꺽정', '장길산', '일지매', '홍길동']
nums = [1, 2, 3, 4, 5, 6, 7, 1]
print(names, type(names))
print(nums, type(nums))
```

```python
names = ('홍길동', '임꺽정', '장길산', '일지매', '홍길동')
nums = (1, 2, 3, 4, 5, 6, 7, 1)
print(names, type(names))
print(nums, type(nums))
```

```python
names = {'홍길동', '임꺽정', '장길산', '일지매', '홍길동'}
nums = {1, 2, 3, 4, 5, 6, 7, 1}
print(names, type(names))
print(nums, type(nums))
```

```python
human = {'홍길동':24, '임꺽정':33, '장길산':44, '일지매': 21}
print(human, type(human))
print(human['홍길동'])
```
